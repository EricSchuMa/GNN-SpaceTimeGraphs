```python
import numpy as np
import pandas as pd
import networkx as nx
import pickle
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%matplotlib inline
```

```python
with open('./data/pems_bay/adj_mx_bay.pkl', 'rb') as f:
    sensor_ids, sensor_id_to_ind, A = pickle.load(f, encoding='latin-1')

dists = pd.read_csv('./data/pems_bay/distances_bay_2017.csv')    
pos = np.load('./data/pems_bay/pos_bay.npy')
```

```python
max_dist = np.max(dists['cost'])
min_dist = np.min(dists['cost'])

dists[['cost']].hist(alpha=0.5, 
                     histtype='bar', 
                     ec='black', 
                     bins=30, 
                     figsize=(20,6),
                    )
plt.title('Distance (km)')
plt.xlabel('Distance between nodes')
plt.show()
```

```python
pd.Series([x for x in A.flatten() if x != 0]).hist(alpha=0.5, 
                     histtype='bar', 
                     ec='black', 
                     bins=30, 
                     figsize=(20,6),
                    )
plt.title('Normalized distance - Containing duplicates and disregarding d(i,j) = 0')
plt.xlabel('Normalized distance between nodes')
plt.show()
```
